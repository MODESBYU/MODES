The following are the spectra of 0M solvent mixtures of various solvent mixtures (ec/emc): 0/100, 20/80, 30/70, 50/50, 70,30, 80/20

```python
B = np.array(pd.read_csv("bkg_dtgs.csv"))[slice(0,3312),1] #Background
wn = np.array(pd.read_csv("spectra_ec_0_emc_100.csv"))[:,0] #Wavenumber set
s01 = np.array(pd.read_csv("spectra_ec_0_emc_100.csv"))[:,1] 
s28 = np.array(pd.read_csv("spectra_ec_20_emc_80.csv"))[:,1]
s37 = np.array(pd.read_csv("spectra_ec_30_emc_70.csv"))[:,1]
s55 = np.array(pd.read_csv("spectra_ec_50_emc_50.csv"))[:,1] 
s73 = np.array(pd.read_csv("spectra_ec_70_emc_30.csv"))[:,1] 
s82 = np.array(pd.read_csv("spectra_ec_80_emc_20.csv"))[:,1] 
```
