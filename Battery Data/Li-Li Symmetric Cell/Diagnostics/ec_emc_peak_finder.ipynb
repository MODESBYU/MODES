The following are the spectra of 0M solvent mixtures of various solvent mixtures (ec/emc): 0/100, 20/80, 30/70, 50/50, 70,30, 80/20

```python
current_path = os.getcwd()
parent_directory = os.path.dirname(current_path)
gparent_directory = os.path.dirname(parent_directory)

folders = gparent_directory+"\\Lydia_data\\Battery_Data\\TQ_calibration\\Single_beam_extracted\\"
folders2 = gparent_directory+"\\Lydia_data\\Battery_Data\\TQ_calibration\\Single_beam_extracted\\Pure EC\\"
```

```python
wn = np.array(pd.read_csv(folders+"spectra_ec_0_emc_100.csv"))[:,0] #Wavenumber set
wn2 =  np.array(pd.read_csv(folders2+"spectra_ec_100_emc_0.tsv",sep="\t"))[:,0]
s01 = np.array(pd.read_csv(folders+"spectra_ec_0_emc_100.csv"))[:,1] 
s28 = np.array(pd.read_csv(folders+"spectra_ec_20_emc_80.csv"))[:,1]
s37 = np.array(pd.read_csv(folders+"spectra_ec_30_emc_70.csv"))[:,1]
s55 = np.array(pd.read_csv(folders+"spectra_ec_50_emc_50.csv"))[:,1] 
s73 = np.array(pd.read_csv(folders+"spectra_ec_70_emc_30.csv"))[:,1] 
s82 = np.array(pd.read_csv(folders+"spectra_ec_80_emc_20.csv"))[:,1] 
s10 = np.array(pd.read_csv(folders2+"spectra_ec_100_emc_0.tsv",sep="\t"))[:,1]
s10_2 = np.array(pd.read_csv(folders2+"spectra_ec_100_emc_0_2.tsv",sep="\t"))[:,1]
s10_3 = np.array(pd.read_csv(folders2+"spectra_ec_100_emc_0_3.tsv",sep="\t"))[:,1]
```

```python
plt.figure(figsize=(20,6))
plt.plot(wn,s01,label="0/100")
plt.plot(wn,s28,label="20/80")
plt.plot(wn,s37,label="30/70")
plt.plot(wn,s55,label="50/50")
plt.plot(wn,s73,label="70/30")
plt.plot(wn,s82,label="80/20")
plt.plot(wn2,s10_3,label="100/0")
plt.legend()
plt.xlim(650,2000)
plt.ylim(0,2)
plt.ylabel("Absorbance")
plt.xlabel("Wavenumber $cm^{-1}$")
#print(len(wn))
plt.title("0M solvent mixtures")
plt.show()
Image(filename="peak_assignments.png",width=600)
```
