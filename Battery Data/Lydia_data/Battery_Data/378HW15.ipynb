```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit
from scipy.optimize import fsolve
from scipy.integrate import odeint, quad, simpson
from scipy.interpolate import interp1d
from scipy.misc import derivative
```

```python
S = np.array([310,223,191,168,153,143,134,127])
N = np.array([2*10**5,1*10**6,3*10**6,1*10**7,3*10**7,1*10**8,3*10**8,1*10**9])
logN = np.log10(N)
plt.plot(logN,S)
plt.xlabel("Cycles to Failure (log(N))")
plt.ylabel("Stress Amplitude, S (MPa)")

fit = np.polyfit(logN,S,2)
strength = np.polyval(fit,np.log10(5*10**5))
print("Fatigue strength at 5*10^5 cycles is ",'%.2f'%strength,"MPa")
print("Fatigue life for 200 MPa is 1.9*10^6 Cycles") 
```
